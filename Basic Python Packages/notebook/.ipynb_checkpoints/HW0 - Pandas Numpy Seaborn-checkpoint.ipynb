#### The iterrows() Method

Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise and values are Python list objects that contain data corresponding to the headers. (Here, use the iterrows method to iterate each row of the dataframe and copy it to a dictionary. However, there is a easier way. Learn how the to_dict() method works by yourself later)






````python
sal_dict ={}
for ids, row in salary_df.iterrows():
    
    #yearID
    if 'yearID' in sal_dict:
        sal_dict['yearID'].append(row['yearID'])
    else:
        sal_dict['yearID'] = [row['yearID']]
    
    #teamID
    if 'teamID' in sal_dict:
        sal_dict['teamID'].append(row['teamID'])
    else:
        sal_dict['teamID'] = [row['teamID']]
        
    #lgID
    if 'lgID' in sal_dict:
        sal_dict['lgID'].append(row['lgID'])
    else:
        sal_dict['lgID'] = [row['lgID']]
    
    #salary
    if 'salary' in sal_dict:
        sal_dict['salary'].append(row['salary'])
    else:
        sal_dict['salary'] = [row['salary']]
    
sal_dict
````

### Dictionary from DF using to_dict

````python
dict = salary_df.to_dict('list')
dict
````

#### Create Dataframe Using the Constructor

Read the documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html#pandas.DataFrame and create a dataframe using pd.DataFrame from the dictionary created in the iterrows() exercise. Change the header to "a", "b", "c", ... at creation time.

````python
dict_df = pd.DataFrame.from_dict(data = sal_dict, orient = 'columns')#.items(),columns = list('ab'))#  {"yearID": "a", "teamID":"b","lgID":"c","salary":"d"})#, columns = ['a','b','c','d'])#, #index=None, columns=None
dict_df.columns = ['a','b','c','d']
# dict_df.rename(columns =  {"yearID": "a",
#                      "teamID":"b","lgID":"c","salary":"d"},
#                                  inplace = True)
dict_df
````

## Pandas's DataFrame.plot

Use the Salaries.csv again (You can use the dataframe object loaded from section 3.1).





a) For team 'ATL', plot a scatter plot between feature yearID and salary.


````python
atl_df = salary_df.loc[salary_df['teamID'] == 'ATL']
atl_df
````

````python
plt.figure(figsize=(10, 8), dpi=80)
plt.title("Salary vs Year of the team ATL")
plt.scatter(atl_df['yearID'], atl_df['salary'])
plt.xlabel('Year')
plt.ylabel('Salary')
plt.yscale("log")
plt.show()
````

(b) For year 1985, plot a bar chart to show the average salary for each team.

````python
yr_df = salary_df.loc[salary_df['yearID'] == 1985]
yr_df
````

````python
plt.figure(figsize=(12, 10), dpi=80)
plt.title("Average salary for each team (1985)")
plt.bar(yr_df['teamID'], yr_df['salary'])
plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
````


(c) For team 'ATL', plot a line chart to show how the annual average salary change by years.

````python
plt.figure(figsize=(12, 10), dpi=80)
plt.title("Changes in annual salary of the ATL team")
plt.plot(atl_df['yearID'], atl_df['salary'], color='green',   
         linewidth=1.0,  
         linestyle='--')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
````
